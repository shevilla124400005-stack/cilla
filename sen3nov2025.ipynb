```python
class Manusia:
    def __init__(self, nama):
        self.nama = nama

class Mahasiswa(Manusia):
    def __init__(self, nama, nim):
        super().__init__(nama)
        self.nim = nim
        
    def nama_nim(self):
        return self.nama, self.nim

manusia2 = Manusia("shevilla")
mahasiswa2 = Mahasiswa("cilla", "123456")
print(manusia2.nama)
print(mahasiswa2.nama)
print(mahasiswa2.nama_nim())
```
